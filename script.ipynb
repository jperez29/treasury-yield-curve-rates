## scraping data

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.treasury.gov/resource-center/data-chart-center/interest-rates/Pages/TextView.aspx?data=yield'
response = requests.get(url).text
soup = BeautifulSoup(response)
table = soup.find("table", attrs={"class": "t-chart"})

#printing the entire table
print(table.prettify())
```

### Printing every row of table

```python
trs = table.findAll("tr")
trs
```

### Getting the names of the columns in table

```python
col_names = table.findAll("th")
column_name = [names.text for names in col_names]
print(column_name)
```

### Printing data on the table, excluding the column names

```python
lst = []
for item in trs:
    td_tag = item.findAll('td')
    for tag in td_tag:
        lst.append(tag.text)
print(lst)
```
