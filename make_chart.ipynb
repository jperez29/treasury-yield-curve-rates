```python
import pandas as pd
import sqlite3 as sql
import numpy as np 
import matplotlib.pyplot as plt
```

```python
db = 'treasury_yield_curve_rates.db'
conn = sql.connect(db)
```

```python

df = pd.read_sql("SELECT * FROM treasury_yield_table", conn)
df
```

```python
df.info()
```

```python
x = df['date']
y = df['one_mo']
fig, ax = plt.subplots(figsize=(20,5))
ax.plot(x,y)
labels = ax.get_xticklabels()
plt.ylim(0,1)
ax.set( xlabel = 'Dates', ylabel= 'curve rates', title = 'some title')
plt.show()
```

```python
x = df['date']
y1 = df['one_mo']
y2=df['two_mo']
y3 = df['three_mo']
plt.plot(x, y1, label = "one_mo")
plt.plot(x,y2, label = "two_mo")
plt.plot(x,y3, label = "three_mo")
plt.legend()
plt.show()
```

```python

import datetime as dt
import numpy as np
import matplotlib.pyplot as plt

def make_chart(data, filename):
    print("generating matplotlib chart")
    plt.plot(data)
#     plt.savefig(f'charts/{filename}.png')
    print("completed")
    
def main():
    dt_now = dt.datetime.now()
    dt_fmt = dt_now.strftime("%m-%d-%y-%H%M%S")
    make_chart(np.arange(10), f'test-{dt_fmt}')
    
main()
```
